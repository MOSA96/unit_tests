# Results


In order to validate that the program is providing the correct results we are comparing the output of the program witht the expected results.

1. Run the program and save its output to a file
```
uv run python main.py files/tests/test_valid.json > actual.out
```

2. Compare results vs expected output

```
diff -u files/results/test_valid_results.txt actual.out
```

No output means the results are the same. Output means that the results differed in one or more line.
